#Data Cleaning

we will fill null values of director with unknown

```python
df['director'].fillna('Unknown',inplace=True)
```

same with cast and country

```python
df['cast'].fillna('Unknown',inplace=True)
```

```python
df['country'].fillna('Unknown',inplace=True)
```

```python
df['date_added'].fillna('Unknown',inplace=True)
```

```python
df['duration'].fillna('Unknown',inplace=True)
df['rating'].fillna('Unknown',inplace=True)
```

```python
df.isnull().sum()
```

check for any duplicate values

```python
df.duplicated().sum()
```

#Univariate analysis

```python
sns.countplot(df['type'])
plt.title('Distribution of movies and tv shows')
plt.show()
```

```python
df['rating'].value_counts().plot(kind='bar')
```

```python
df['rating'].unique()
```

```python
df = df[~df['rating'].str.contains('min', na=False)]
```

```python
df['rating'].unique()
```

```python
df['rating'].value_counts().plot(kind='bar')
```

```python
sns.countplot(df['rating'])
```

top 10 countries

```python
df.groupby('country')['title'].count().sort_values(ascending=False).head(10)
```

```python
(df.groupby('country')['title'].count().sort_values(ascending=False).head(10)).plot(kind='bar')
```

```python
yearcounts=df.groupby('release_year')['title'].count().sort_values(ascending=False).head(10)
```

```python
plt.bar(yearcounts.index,yearcounts.values)
plt.title('Number of netflix releases per year')
plt.xlabel('Year')
plt.ylabel('Number of releases')
plt.show()
```
